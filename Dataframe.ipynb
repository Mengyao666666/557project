```python
df_skew = pd.read_csv("/content/drive/MyDrive/557project/CBOE_SKEW_Full_2000_2026.csv")
df_vix = pd.read_csv("/content/drive/MyDrive/557project/VIX.csv")
df_spread = pd.read_csv("/content/drive/MyDrive/557project/spread_10y2y.csv")
df_oas = pd.read_csv("/content/drive/MyDrive/557project/hy_oas.csv")
```

```python
# 定义一个列表方便批量操作
dataframes = {
    "SKEW": df_skew,
    "VIX": df_vix,
    "10y-2y Spread": df_spread,
    "HY OAS": df_oas
}

for name, df in dataframes.items():
    print(f"--- {name} ---")
    print(f"行列维度 (Rows, Cols): {df.shape}")
    print("前 3 行预览:")
    display(df.head(3)) # 在 Colab/Jupyter 里用 display 更好看
    print("\n")
```

```python
df_vix = df_vix.iloc[2:].copy()

df_vix.set_index('Price', inplace=True)
df_vix.index.name = 'Date'

print(df_vix.head())
```

```python
print(df_skew.head())
```

```python
for df in [df_spread, df_oas]:
    df.columns = df.columns.str.strip()
```

```python
for df in [df_spread, df_oas]:
    df['observation_date'] = pd.to_datetime(df['observation_date'])
    df.set_index('observation_date', inplace=True)
```

```python
print(df_spread.head())
print(df_oas.head())
```

```python
for df in [df_skew, df_vix]:
    df.index = pd.to_datetime(df.index)
```

```python
data_final = pd.concat([
    df_skew['Close'].rename('SKEW'),
    df_vix['Close'].rename('VIX'),
    df_spread['T10Y2Y'].rename('Spread'),
    df_oas['BAMLH0A0HYM2'].rename('OAS')
], axis=1)
```

```python
print(data_final.head())
print(data_final.shape)
```

```python
print("原始 shape:", data_final.shape)
print("dropna 后 shape:", data_final.dropna().shape)
print("\n每列缺失数：")
print(data_final.isna().sum())
```

```python
data_model = data_final.dropna().copy()
print(data_model.shape)
```
